 12. Classify the employees into 3 categories.
 Category 1 : employees below 18.5 bmi - underweight
 Category 2 : employees with bmi in the range of 18.5 to 24.9 - fit
 Category 3 : employee with bmi over 24.9 - overweight
 df1


```python
underweight=df1[df1.bmi<18.5]
fit=df1[(df1.bmi>=18.5) & (df1.bmi<=24.9)]
overweight=df1[df1.bmi>24.9]
A=pd.DataFrame(underweight)
B=pd.DataFrame(fit)
C=pd.DataFrame(overweight)
print('average expenditure of employee in each category')
print('underweight:',round(A.expenditure.mean(),2))
print('fit:',round(B.expenditure.mean(),2))
print('overweight:',round(C.expenditure.mean(),2))
plt.bar(x=['underweight','fit','overweight'],height=[A.expenditure.mean(),B.expenditure.mean(),C.expenditure.mean()],color=['pink'])
plt.xlabel('categories')
plt.ylabel('average expenditure')
plt.show()
```

 13. Draw line plot of expenditure against age and BMI

```python
plt.figure()
a=df1.sort_values('age')
b=df1.sort_values('bmi')
c=df1.sort_values('expenditure')
plt.plot(a.age,df1.expenditure,color='pink',marker=">")
plt.plot(b.bmi,df1.expenditure,color='red',marker='*')
size=plt.gcf()
size.set_size_inches(12,6)
plt.title('line plot of expenditure against age & bmi ')
plt.show()
```
